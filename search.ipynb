```python
# đọc file 1 dòng trong file json và chỉ lấy những field mà chúng ta quan tâm
def parse_fields(line):
    data = json.loads(line)
    return {
        "book_id": data["book_id"],
        "title": data["title_without_series"],
        "ratings": data["ratings_count"],
        "url": data["url"],
        "cover_image": data["image_url"]
    }
```

```python
#tạo danh sách những cuốn sách được recommend từ file json

books_titles = []
with gzip.open("goodreads_books.json.gz", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        fields = parse_fields(line)
        #convert ratings to int
        try:
            ratings = int(fields["ratings"])
        except ValueError:
            continue
        if ratings > 15:
            books_titles.append(fields)
```

```python
import pandas as pd
# tạo titles data frame chứa những cuốn sách được recommend
titles = pd.DataFrame.from_dict(books_titles)
```

```python
titles["ratings"] = pd.to_numeric(titles["ratings"])
```

```python
#thêm 1 cột "mod_titlé"trong data  frame là title sau khi được modified
titles["mod_title"] = titles["title"].str.replace("[^a-zA-Z0-9 ]", "",regex=True)
```

```python
#data frame output sau khi xử lí|
titles
```

```python
# huyển toàn bộ titles về kí tự thường 
titles["mod_title"] = titles["mod_title"].str.lower()
# chuyển  
```

```python
titles["mod_title"] = titles["mod_title"].str.replace("\s+", " ", regex=True)
```

```python
#loai bo những sách có mod_title. length == 0
titles = titles[titles["mod_title"].str.len() > 0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
#tfidf chứa các ma trận sau khi chuyển các "mod_titles" thành dạng ma trận 
tfidf = vectorizer.fit_transform(titles["mod_title"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import re

def make_clickable(val):
    return '<a target="_blank" href="{}">Goodreads</a>'.format(val)
def show_image(val):
    return '<img  src="{}" with=50></img>'.format(val)

#query chứa từ khóa tìm kiếm, vectorizer chứa ??
def search(query, vectorizer):
    #query = "East"
    #xử lí từ khóa thành dạng đơn giản
    processed = re.sub("[^a-zA-Z0-9]", "", query.lower())
    #biến đổi từ khóa bằng thuật toán
    query_vec = vectorizer.transform([processed])
    similarity = cosine_similarity(query_vec,tfidf).flatten()
    #tìm 10 hàng có giá trị từ gần giống với từ khóa dựa vào ma trận
    indices = np.argpartition(similarity, -10)[-10:]
    results = titles.iloc[indices]
    results = results.sort_values("ratings", ascending=False)
    #lấy 5 hàng có gần giống nhất với từ khóa đã cho
    return results.head(5).style.format({'url': make_clickable, 'cover_image' : show_image})
```

```python
search("king", vectorizer)
```
