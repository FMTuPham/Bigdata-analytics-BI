```python
liked_books = ["2517439", "113576","356824"]
```

```python
csv_book_mapping = {}

with open("book_id_map.csv", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        csv_id, book_id = line.strip().split(",")
        csv_book_mapping[csv_id] = book_id
```

```python
len(csv_book_mapping)
```

```python
overlap_users = set()
# tìm users có đọc những cuốn sách nằm trong liked_books 
with open("goodreads_interactions.csv", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        try:
            user_id, csv_id, _, rating, _ = line.split(",")
        except ValueError:
            continue
        if user_id in overlap_users:
            continue
        
        try:
            rating = int(rating)
        except ValueError:
            continue
        book_id = csv_book_mapping[csv_id]
        
        if book_id in liked_books and rating>=4:
            overlap_users.add(user_id)
        
```

```python
overlap_users
```

```python
rec_lines = []
# tạo danh sách những cuốn sách khác mà users nằm trong overlap_users đọc
with open("goodreads_interactions.csv", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        try:
            user_id, csv_id, _, rating, _ = line.split(",")
        except ValueError:
            continue
        if user_id in overlap_users:
            book_id = csv_book_mapping[csv_id]
            rec_lines.append([user_id, book_id, rating])
```

```python
import pandas as pd
#tạo dataFrame cho recs_line
recs = pd.DataFrame(rec_lines, columns=["user_id", "book_id", "rating"])
recs["book_id"] = recs["book_id"].astype(str)
```

```python
#10 cuốn sách được đọc nhiều nhất
top_recs = recs["book_id"].value_counts().head(10)
top_recs = top_recs.index.values
```
